```python
#import all the neccessary libraries
import numpy as np #help in mathematics
import pandas as pd #help in data import and cleaning
import matplotlib.pyplot as plt # helps in visualising the graph

dataset = pd.read_csv('/content/ML-hands-on/Part 2 - Regression/Section 4 - Simple Linear Regression/Salary_Data.csv')
```

```python
#saving independent and dependent variables
X = dataset.iloc[:, :-1].values #all col,except for last(: means all)
y = dataset.iloc[:, -1].values #last col(-1 means last)
```

```python
# Spliting  the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Train the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #Creating the class instance for LR
regressor.fit(X_train, y_train) # using methood called fit , most important line
```

Predictions

```python
y_pred = regressor.predict(X_test)
```

Visualization

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
